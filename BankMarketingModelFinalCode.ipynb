**Technique 1: Impute with most frequent value**

```python
# for column in unknowns[unknowns].index:
#     most_frequent = final_df[column].mode()[0]
#     print(most_frequent)
#     final_df[column] = final_df[column].replace('unknown', most_frequent)

from scipy import stats
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier


def impute_most_frequent(df, columns_with_unknown):
    df_imputed = df.copy()
    for column in columns_with_unknown:
          if 'unknown' in df_imputed[column].values:
              imputer = SimpleImputer(strategy='most_frequent')
              df_imputed[column] = imputer.fit_transform(df_imputed[[column]]).ravel()
    return df_imputed
```

**Technique 2: Impute using Randomforest model**

```python

# Function to impute missing values using RandomForestClassifier
def impute_random_forest(df, columns_with_unknown):
    df_imputed = df.copy()

    for column in columns_with_unknown:
        if df_imputed[column].isna().any():
            known_data = df_imputed[df_imputed[column].notna()]
            unknown_data = df_imputed[df_imputed[column].isna()]

            X_train = known_data.drop(columns=[column])
            y_train = known_data[column]

            X_test = unknown_data.drop(columns=[column])

            model = RandomForestClassifier(n_estimators=100, random_state=42)
            model.fit(X_train, y_train)

            predicted_values = model.predict(X_test)
            df_imputed.loc[df_imputed[column].isna(), column] = predicted_values

    return df_imputed
```

**comaparison of both method**

```python
# Function to perform KS test
def ks_test(df1, df2, columns_with_unknown):
    ks_stats = {}

    for column in columns_with_unknown: # Iterate over the list
        ks_stat, p_value = stats.ks_2samp(df1[column].dropna(), df2[column].dropna())
        ks_stats[column] = (ks_stat, p_value)

    return ks_stats

# Impute missing values using different methods
df1 = final_df.copy()
df2 = final_df.copy()
imputed_most_frequent = impute_most_frequent(df1, columns_with_unknown)
imputed_rf = impute_random_forest(df2, columns_with_unknown)

# Perform KS test to compare distributions
ks_stats = ks_test(imputed_most_frequent, imputed_rf, columns_with_unknown)

# Print KS statistics and p-values
for column, (ks_stat, p_value) in ks_stats.items():
    print(f"Column: {column}, KS Statistic: {ks_stat}, p-value: {p_value}")
```

**Technique 1 outlier handling using Z score**

```python
df3 = final_df[numerical_columns].copy()
# Function to detect outliers using Z-score
def detect_outliers_zscore(df, threshold=3):
    outliers = pd.DataFrame()

    for column in df.columns:
        z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
        column_outliers = df[z_scores > threshold]
        column_outliers['Outlier_ZScore_' + column] = z_scores[z_scores > threshold]
        outliers = pd.concat([outliers, column_outliers], axis=1)

    return outliers

# Detect outliers using Z-score method
outliers_zscore = detect_outliers_zscore(df3)

print("Outliers detected using Z-score:")
print(outliers_zscore)
```

**Technique 2 outlier handling using IQR**

```python
df4 = final_df[numerical_columns].copy()
# Function to detect outliers using IQR
def detect_outliers_iqr(df, threshold=1.5):
    outliers = pd.DataFrame()

    for column in df.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        column_outliers = df[(df[column] < Q1 - threshold * IQR) | (df[column] > Q3 + threshold * IQR)]
        column_outliers['Outlier_IQR_' + column] = True
        outliers = pd.concat([outliers, column_outliers], axis=1)

    return outliers

# Detect outliers using IQR method
outliers_iqr = detect_outliers_iqr(df4)

print("\nOutliers detected using IQR:")
print(outliers_iqr)
```
